```python
n=int(input("Enter number of students in the class: "))
present_count=0
absent_count=0
rollno=1
while rollno<=n:
    print("enter roll no",rollno,"is present or absent select following options 1 0r 2/n 1. Present/n 2. Absent:")
    status=input()
    if status=='1':
        present_count+=1
        rollno+=1
    elif status=='2':
        absent_count+=1
        rollno+=1
    else:
        print("please select either 1 or 2 option only")
print("Total number of students present in the class:",n)
print("Total number of students present in the class:",present_count)
print("Total number of students absent in the class:",absent_count)
percentage=(present_count/n)*100
print("Percentage of students present in the class:",percentage,"%")
```
