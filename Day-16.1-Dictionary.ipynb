```python
#Access Dictionary values
student={'name':'krishna','age':21,'Bach':47,'course':'Python'}
print(student['name'])
print(student['Name']) #if key is not in dictionary it raises key error
```

```python
#Updating the dictionary values and items
student={'name':'krishna','age':21,'Bach':47,'course':'Python'}
student['name']='Naveen'
print(student['name'])
```

```python
student={'name':'krishna','age':21,'Bach':47,'course':'Python'}
student['name']='Naveen'
print(student['name'])
#Assign new item into dictionary
#dict_obj[key]=value
student['year']=2005
print(student)
```

```python
#Dictionary Functions():
#Accesing values
#get(key,default_value):it return value if key is present in dict otherwise default values
student={'name': 'Naveen', 'age': 21, 'Bach': 47, 'course': 'Python', 'year': 2005}
print(student.get('name'))
print(student.get('percentage'))
print(student.get('percentage',90))
```

```python
#Updating or Adding items to dictionary
#Updates(new_dictionary):It updates with new_dictionary items
student={'name':'Krishna','batch':47,'course':'python','age':21,'percentage':80}
marks={'python':90,'soft kills':85,'apptitude':89,'percentage':90}
student.update(marks)
print(student)
print(marks)
```

```python
#Dictionary not allow duplicates keys
student={'name':'Krishna','batch':47,'course':'python','age':21,'percentage':80,'percentage':90,'name':'sai'}
print(student)
```

```python
#get dictionary keys and values
#keys()-It returns list of all dictionary keys
#values()-It returns list of all dictionary values
#Items()-It returns all dictionary items as list od tuples of key-values pairs
student={'name':'Krishna','batch':47,'course':'python','age':21,'percentage':80,'percentage':90,'name':'sai'}
print(student.keys())
print(student.values())
print(student.items())
```

```python
student={'name':'Krishna','batch':47,'course':'python','age':21,'percentage':80,'percentage':90,'name':'sai'}
print(type(student.keys()))
print(type(student.values()))
print(type(student.items()))
```

```python
#dictionary-Type coversion 
student={'name':'Krishna','batch':47,'course':'python','age':21,'percentage':80,'percentage':90,'name':'sai'}
print(list(student.keys()))
print(list(student.values()))
print(list(student.items()))

print(tuple(student.keys()))
print(tuple(student.values()))
print(tuple(student.items()))

print(str(student.keys()))
print(str(student.values()))
print(str(student.items()))

print(set(student.keys()))
print(set(student.values()))
print(set(student.items()))
```

```python
#Deleting items from dictionary
#pop(key)-It returns and removes the specified key items from dictionary
#popitem()-It returns and removes the last items from dictionary
student={'name':'Krishna','batch':47,'course':'python','age':21}
poped_value=student.pop('age')
print(poped_value)
print(student)
poped_value=student.pop('age',None)
print(poped_value)
print(student)
items=student.popitem()
print(items)
print(student)
```

```python
#Iterate dictionary items 
student={'name':'Krishna','percentage':90,'batch':47,'course':'python','age':21,'percentage':80}
for item in student.items():
    print(item)
```

```python
student={'name':'Krishna','percentage':90,'batch':47,'course':'python','age':21,'percentage':80}
for item in student.items():
    print(f"{item[0]}:{item[1]}")
```

```python
student={'name':'Krishna','percentage':90,'batch':47,'course':'python','age':21,'percentage':80}
for keys,values in student.items():
    print(f"{keys}:{values}")
```
