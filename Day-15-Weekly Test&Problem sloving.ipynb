```python
#Finding length of number
#Ex-456 o/p-3
#15624 o/p-5
num=int(input("Enter a number:"))
count=0
while num!=0:
    num=num//10
    count+=1
print("length of number:",count)
```

```python
#Sum of digits in a number
n=int(input("enter a number:"))
s=0
while n>0:
    digit=n%10
    n=n//10
    s=s+digit
print(s)
```

```python
#Reverse Numbers :
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reverse of the number is:", rev)
```

```python
#Amstrong number
num = int(input("Enter a number: "))

temp = num
sum = 0

# find number of digits
count = 0
temp2 = num
while temp2 > 0:
    count += 1
    temp2 //= 10

# calculate armstrong sum
while temp > 0:
    digit = temp % 10
    sum += digit ** count
    temp //= 10

if sum == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```
