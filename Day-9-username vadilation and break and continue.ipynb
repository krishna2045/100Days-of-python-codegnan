```python
#password validation
password=input("enter a password")
upper=lower=digit=special=0
if len(password)>8:
    for ch in password:
        if ch.isupper():
            upper = 1
        elif ch.islower():
            lower = 1
        elif ch.isdigit():
            digit = 1
        elif ch in "@#$%&":
            special=1
    if upper and lower and digit and special:
        print("password is valid")
    else:
        print("password is invalid")
else:
    print("password must be at least 8 character long")
```
