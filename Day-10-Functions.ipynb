```python
#FUNCTIONS()-Fn is a block of reusable code that perform multiple tasks and runs only when its is called "functions"
def sum_of_two_numbers(a,b): #fn definition
    s=a+b
    return s
result=sum_of_two_numbers(10,20)
print(result)
```

```python
#calculator program using function:
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return(a*b)
def div(a,b):
    if b !=0:
        return a/b
    else:
        return "division by zero is not possible"
#Taking input from user
num1=int(input("enter a number:"))
num2=int(input("enter a number:"))
print("select operations")
print("1.add")
print("2.sub")
print("3.mul")
print("4.div")
choice=int(input("enter your choice(1/2/3/4):"))
#using functions
if choice==1:
    print("result:",add(num1,num2))
elif choice==2:
    print("reult:",sub(num1,num2))
elif choice==3:
    print("result:",mul(num1,num2))
elif choice==4:
    print("result:",div(num1,num2))
else:
    print("Invalid choice")
```

```python
#EVEN OR ODD:
def even_or_odd(num):#function definition
    if num%2==0:
        return True
    else:
        return False

number=int(input("enter a number:"))
result=even_or_odd(number)       #Function calling
if result==True:    
    print("is even")
else:
    print("is odd")
```

```python
#rread list of number and if number is even add 0.5 and if number is odd add 1
lst=list(map(int,input().split()))
for i in range(0,len(lst)):
    if lst[i]%2:
        lst[i]+=1
    else:
        lst[i]+=0.5
    print(lst)
```
