```python
#Operations
```

```python
#Union
s1={1,2,3,4,5}
s2={5,4,6,78,9}
print(s1.union(s2)) #or use (s1|s2)
```

```python
#Intersection
s1={1,2,3,4,5}
s2={5,4,6,78,9}
print(s1.intersection(s2)) # S1 & S2
```

```python
#Difference:
s1={1,2,3,4,5,10}
s2={5,4,6,78,9}
print(s1.difference(s2)) # s1-s2
```

```python
#Symmetric_difference
s1={1,2,3,4,5}
s2={5,4,6,78,9}
print(s1.symmetric_difference(s2)) #s1 ^ s2
```

```python
#issubset()
s1={1,2,3,4,5,10}
s2={5,4,6,78,9}
print(s1.issubset(s2)) # s1-s2
print({1}.issubset(s1))
```

```python
s1={1,2,3,4,5,10}
s2={5,4,6,78,9}
print(s1.isdisjoint(s2)) # s1-s2
print({1}.isdisjoint(s1))
```

```python
#isdisjoint
={1,2,3,4,5}
s2= {7, 8}
print(s1.isdisjoint(s2))
```

```python
#issuperset()
s1={1,2,3,4,5,10}
s2={5,4,6,78,9}
print(s1.issuperset(s2))
```

```python
s1={1,2,3,4,5}
s2={1,2,3,4,5}
print(s1.issuperset(s2))
```
