```python
#2.default aruguments
def add(a,b,c=0,d=0):
    res=a+b+c+d
    return res
print(add(1,2,3,4))
print(add(1,2,3))
print(add(1,2))
```

```python
#using default arguments 2,3,4-numbers multiplication:
def mul(a,b,c=1,d=1):
    res=a*b*c*d
    return res
print(mul(1,2,3,4))
print(mul(1,2,3))
print(mul(1,2))
```

```python
#long numbers we cant use default arguments so we overcome that use variable length positional arguments...
#ariable length positional arguments:
a,*b=10,20,30,40,50
print(a)
print(b)
```
